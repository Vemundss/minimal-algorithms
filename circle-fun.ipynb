```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
freqs = [20,1,1]
scales = [0,0.2,1,1]
theta = np.linspace(0,np.pi*2,1000)
x = scales[0] + sum([scales[i+1]*np.sin(freqs[i]*theta) + 
                     scales[i+1]*np.cos(freqs[i]*theta) for i in range(len(freqs))])
y = scales[0] + sum([scales[i+1]*np.sin(freqs[i]*theta+np.pi/2) +
                     scales[i+1]*np.cos(freqs[i]*theta+np.pi/2) for i in range(len(freqs))])
```

```python
plt.plot(x,y)
```
