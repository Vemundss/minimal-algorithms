```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(3, 3))
t = np.linspace(0, 2*np.pi, 1000)
x = np.sin(t)
y = np.cos(t)
mask = (x >= 0) & (y >= 0)

plt.plot(x[mask], y[mask], lw=0.5, label='activity (norm + relu)')
plt.plot(x[~mask], y[~mask], lw=0.5, ls='--')
plt.plot(np.linspace(1, 0, 2), lw=0.5, label='L1 of activity')
plt.plot(x[mask]*(x[mask] + y[mask]), y[mask]*(y[mask] + x[mask]), lw=0.5, label='L1 of activity (norm + relu)')
plt.axis('equal')
plt.scatter(1, 0, c='r', label='L1-minima')
plt.scatter(0, 1, c='r')
plt.scatter(1/np.sqrt(2), 1/np.sqrt(2), c='b', label='L1-maxima')
plt.legend()
```
